```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic graph

```python
#normally x and y are 1d arrays
plt.plot([1,2,3],[4,5,6])
```

```python
#or
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y)
#plt.show() gets rid of this[<matplotlib.lines.Line2D at 0x22b61a3d4c8>] on the graph in jupyter notebook
```

```python
#add title
#add laels to the axis
#fontname  and fontsize can be added to the title(pass the fontname and fontsize as a dictionary
#the labels can also have different font names and sizes[ensure you know the fontname])
#by using xticks and yticks..the graph displays the exact data we have entered in the x and y axis(no floats or conversion)
#label can be added to the plot or passed into the legend
#color can be passed intpo the plot function to change colours..they can also be shortened,color for c,blue for b green for g and so on
#we can add a line width to the plot function
#marker can be added..(check the documentation to see the markers and lines supported )in the graph and the size can be increased with marker size
#markeredgecolor can change the colour of the marker..it can be shortened or written in full
#linestyle can also be passed (you can read the matplotlib documentation to get more)
#there is also a shorthand way of writing a shorthand instead of listing all the plot functions
#fmt = "[color][line][marker]"
x = [0,2,3,4,5]
y = [0,4,6,8,10]
#plt.plot(x,y, label= "2x",color="red",linewidth = 4,linestyle = "--", marker = "*",markersize = 12,markeredgecolor = "blue" )
#shorthand
#fmt = "[color][marker][line]" as a letter of the assignment and as a dot product

plt.plot(x,y,"r*--",label = "2x")#the arguments are passed in directly as a string in that format
plt.title("Basic line graph",fontdict={"fontname" :"Calibri", "fontsize":24})
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.xticks([1,2,3,4,5])
plt.yticks([2,4,6,8,10])
plt.show()
```

```python
#this code will display the font names available in the matplotlib for your computer
import matplotlib.font_manager

from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
#incase some numbers are entered and they are not straight forward.such as sine we can do the following
plt.plot(x,y,"r*--",label = "2x")
x1 = np.arange(0,4.5,0.5)#array between 0 and 4(excluding 4)ference of 0.4
x2 = x1**2
plt.plot(x1,x2,label="x*2")
plt.legend()
plt.show()
```

```python
#to exagerate different parts of the line.. I,E PLOT PART OF THE GRAPH AS A LINE.we can do that by slicing
#we plot rows differently and plot columns differently
x1 = np.arange(0,4.5,0.5)#array between 0 and 4(excluding 4)ference of 0.4
plt.plot(x,y,"r*--",label = "2x")
plt.plot(x1[:6],x1[:6]**2,label="x*2")
plt.plot(x1[5:],x1[5:]**2,"b--",label="x*2")
plt.legend()
plt.show()
```

```python
#to resize the graph we do as follows
x1 = np.arange(0,4.5,0.5)#array between 0 and 4(excluding 4) withspace inbetween each [point as 0.5]
plt.figure(figsize =(5,2.5),dpi =300) # dpi means the pixels by inch.. the large the deeper
plt.plot(x,y,"r*--",label = "2x")
plt.plot(x1[:6],x1[:6]**2,label="x*2")
plt.plot(x1[5:],x1[5:]**2,"b--")
plt.legend()
plt.show()
```

# bar chart

```python
#to construct bar chart
labels= ["a","b","c"]
values = [23,35,46]
plt.bar(labels,values)
plt.show()
```

```python
#we hatch to different labels= ["a","b","c"]
#we can also plot figure size
values = [23,35,46]
bar_chart = plt.bar(labels,values)
plt.figure(figsize=(6,3))
bar_chart[0].set_hatch("/")
bar_chart[1].set_hatch("*")
bar_chart[2].set_hatch("-")# this sets the hatch design on each bar by index
plt.show()
```

```python
#if there are many bars,setting the hatch for each bar could be stressful so we can use a for loops as folows
values = [23,35,46]
bar_chart = plt.bar(labels,values)
plt.figure(figsize=(6,3))
patterns = ["/","*","-"]
for bar in bar_chart:
    bar.set_hatch(patterns.pop(0)) #we set the pop to 0 so it starts from the first bar if not it atarts from the last
plt.show()
```
